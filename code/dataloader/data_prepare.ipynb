```python
import os
import glob
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
import h5py

def get_bbox(gt):
    mask = (gt != 0)
    brain_voxels = np.where(mask != 0)
    minXidx = int(np.min(brain_voxels[0]))
    maxXidx = int(np.max(brain_voxels[0]))
    minYidx = int(np.min(brain_voxels[1]))
    maxYidx = int(np.max(brain_voxels[1]))
    return minXidx, maxXidx, minYidx, maxYidx
    
def center_crop(img, lab, crop_size):
    minXidx, maxXidx, minYidx, maxYidx = get_bbox(lab)
    print(maxXidx - minXidx, maxYidx - minYidx)
    x_extend = (crop_size[0] - (maxXidx - minXidx)) // 2
    y_extend = (crop_size[1] - (maxYidx - minYidx)) // 2
    
    new_minXind = max(0, minXidx - x_extend)
    new_maxXind = min(new_minXind+crop_size[0], img.shape[0])
    
    new_minYind = max(0, minYidx - y_extend)
    new_maxYind = min(new_minYind+crop_size[1], img.shape[1])
    
    return img[new_minXind:new_maxXind, new_minYind:new_maxYind], lab[new_minXind:new_maxXind, new_minYind:new_maxYind]
    
    
raw_path = "D:/SMU-GC/dataset/roi1 540/*"

raw_path_lists = glob.glob(raw_path)
for i in raw_path_lists:
    print(i.split("\\"))
    reader = sitk.ImageSeriesReader()
    dicom_names = reader.GetGDCMSeriesFileNames(i)
    reader.SetFileNames(dicom_names)
    image = reader.Execute()
    image_arr = sitk.GetArrayFromImage(image)
#     image_arr[image_arr > 275] = 275
#     image_arr[image_arr < -125] = -125
    label = sitk.GetArrayFromImage(sitk.ReadImage(i + "/Untitled.nii"))
    
#     f = h5py.File('D:/SMU-GC/dataset/SMU/{}.h5'.format(i.split("\\")[-1]), 'w')
#     f.create_dataset('image', data=image_arr[0, ...], compression="gzip")
#     f.create_dataset('label', data=label[0, ...], compression="gzip")
#     f.close()
    
#     cropped_img, cropped_lab = center_crop(image_arr[0, ...], label[0, ...], (160, 160))
#     plt.imshow(cropped_img, cmap="gray")
#     plt.show()
    
#     plt.imshow(cropped_lab, cmap="gray")
#     plt.show()
```
